# Various Weight nitialization Techniques

## 1. Uniform Distribution

#### The formula is:
> (-1/sqrt(fan_in),1/sqrt(fan_in))

```python
Samples=np.array(np.random.randn(10,3))
N=Samples.shape[1]
```

```python
weights_Uniform=np.random.uniform((-1/np.sqrt(N)),(1/np.sqrt(np.sqrt(N))),1000)
```

### Function to initialize weights

```python
def uniform(No_of_Input,row,col):
    return np.random.uniform((-1/np.sqrt(No_of_Input)),(1/np.sqrt(np.sqrt(No_of_Input))),(row,col))
```

```python
# Let's plot the weights
```

```python
sns.distplot(weights_Uniform,label='Uniform Distribution')
```

> 1. We can see that weights are distributed normally
> 2. By this way we can avoid getting duplicate values for weights 
> 3. This method works verywell with Sigmoid function

# 2. Xavier or Glorat

## 1. Xavier or Goret Normal
> 1. The weights are taken from a Normal distribution where mean is 0 and some standard deviation:
>> standartd deviation = sqrt(2/Fan_in+fan_out)

```python
X=np.array(np.random.randn(10,3))
Hidden_layer=4
```

```python
weights_Xavier=np.random.normal(0,np.sqrt(2/((X.shape[1])+(Hidden_layer))),1000)
```

### Function to initialize weights

```python
def glorat_normal(no_input,No_output):
    return np.random.normal(0,np.sqrt(2/(no_input+No_output)),(no_input,No_output))
```

```python
# Let's plot the weights
```

```python
sns.displot(weights_Xavier)
```

# Points to Note:
>1. The probability of a weight to be in between -0.4 to 0.4 is high.
>2. by this way we can keep our weight not to be very big or very small
